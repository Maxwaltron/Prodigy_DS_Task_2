```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/drive/MyDrive/Project_Datasets/Titanic_Dataset/train.csv")
data.head()  # View the top rows
```

```python
data.info()  # Data types and non-null counts
data.describe()  # Statistical summary for numerical columns
```

```python
data.isnull().sum()  # Check for missing values in each column
```

```python
df=pd.DataFrame(data)
```

```python
df.drop(columns='Cabin').isna().mean()
df.drop(columns='Cabin').dropna(subset=['Embarked'])
```

```python
df.dropna(subset=['Age'],inplace=True)
df.drop('Cabin', axis=1, inplace=True)
df
```

```python
df.to_csv("/content/drive/MyDrive/Project_Datasets/Titanic_Dataset/train_Cleaned.csv")
```

```python
sns.histplot(data=data, x="Age", hue="Survived", multiple="stack")
```

```python
survived_female_counts = data[data['Sex'] == 'female']['Survived'].value_counts()
print(f"Survived: {survived_female_counts[1]}")
print(f"Not Survived: {survived_female_counts[0]}")
```

```python
survived_male_counts = data[data['Sex'] == 'male']['Survived'].value_counts()
print(f"Survived: {survived_male_counts[1]}")
print(f"Not Survived: {survived_male_counts[0]}")
```

```python
data['Survived'].value_counts()  # Counts of 'Survived' and 'Not Survived'
sns.countplot(x='Pclass', data=data)  # Bar plot of passenger class distribution
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Ticket'].value_counts()
    for x_label, grp in df.groupby('Sex')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Sex')
_ = plt.ylabel('Ticket')
```
